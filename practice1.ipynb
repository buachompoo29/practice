**(1a) Find the maximum value and display the maximum value to the console.**

```python
alist = [1,2,3,4,5]
max = alist[0]
for i in range(len(alist)):
  if alist[i] > max:
    max = alist[i]

print('maximum = ', max)
```

(2b) Count the number of even numbers and the number of odd numbers, and then display both
numbers to the console. For example, if the list of data contains the number 1,2, 3, 4 and 5,
there will be 2 even numbers and 3 odd numbers.

```python
counteven = 0
countodd = 0
for i in range(len(alist)):
  if alist[i] % 2 == 0:
    counteven +=1
  elif alist[i] % 2 == 1:
    countodd +=1
print('the number of even numbers = ',counteven)
print('the number of odd numbers = ',countodd)
```

(2) Write a function to calculate the body mass index (BMI) by taking the weight and height as
inputs. If the BMI is greater than 30, display the string "Your BMI is too much." If the BMI is less
than or equal to 30, display the string "Your BMI is normal." The BMI in the string should be a
float value calculated by the program. To call the function, you can prompt the user for their
weight and height and store them in variables. You can then pass these variables as arguments
to the function. For example, if the input is a male with a weight of 80 kilograms and a height of
1.68 meters, the program should display the result as shown in the image.

```python
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi
def eval(bmi):
  if bmi > 30:
    return "too much"
  else:
    return "normal"

weight = float(input("Enter your weight: "))
height = float(input("Enter your height: "))

bmi = calculate_bmi(weight, height)
eval_result = eval(bmi)
print(f"Your {bmi} is {eval_result}.")
```

(3) Write a program to receive an English sentence from the keyboard. Then, translate every vowel
in the English sentence (a, e, i, o, u) into an *, regardless of whether it is uppercase or
lowercase. For example, if the English sentence "Financial Programming " is entered, the result
will be as follows

```python
text = input("Enter your string: ")
vowel = ['a','e','i','o','u']
text_list = list(text)
for i in range(len(text_list)):
  if text_list[i] in vowel:
    text_list[i] = '*'
new_text = ''.join(text_list)
print(new_text)
```

(4) Write a program to count the vowels in the given English phrase. Declare a dictionary variable,
v= {'a':0,'e':0,'i': 0,'o':0,'u':0}, to use for counting vowels. If the phrase "Financial Programming"
is entered, the result will be as follows

```python
text = input("Input your sentence: ").lower()
v= {'a':0,'e':0,'i': 0,'o':0,'u':0}
vowel = ['a','e','i','o','u']
for i in text:
  if i in vowel:
    v[i] += 1
print(v)
```

(5) Scrape website

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.fruitnflora.com/flower-bouquets'

r = requests.get(url)

data = r.text
print(data)

soup = BeautifulSoup(data, 'html.parser')

product_name = []
price = []
link =[]
```

```python
product_name.clear()
print(product_name)
price.clear()
print(price)
link.clear()
print(link)
```

```python
for product in soup.find_all('div', class_="ProductName"):
    product_name.append(product.get_text(strip=True))


print(product_name)
```

```python
for p in soup.find_all('div', class_="ProductPriceRating"):
    price.append(p.get_text(strip=True))

print(price)
```

```python
for l in soup.find_all('div', class_="ProductName"):
    a_tag = l.find('a')
    if a_tag:
        link.append(a_tag.get('href'))
print(link)
```
